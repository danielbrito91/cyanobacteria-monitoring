```python
# Hard code cloud imgs
clusters_fts.loc[
    (pd.to_datetime(clusters_fts["date"])
    .dt.date
    .isin(
        pd.to_datetime(
            config["gee_clean"]["imgs_com_nuvens"])
            .date)),
        "labels"] = config["gee_clean"]["kmeans"]["clean_cluster"] + 1

gee_train = clusters_fts .loc[pd.to_datetime(gee["date"]) < pd.to_datetime('2021-07-01')]
gee_test = clusters_fts.loc[pd.to_datetime(gee["date"]) > pd.to_datetime('2021-07-01')]

X_train = gee_train.iloc[:, 2:]
X_test = gee_test.iloc[:, 2:]
y_train =( gee_train["labels"] == 0) * 1
y_test = (gee_test["labels"] == 0) * 1
```

```python
clst = clean_gee.cluster_gee((config))
X_train, X_test, y_train, y_test = clean_gee.split_data_cloud(clst, 0)

cloud_clf = clean_gee.train_cloud_classifier(X_train, y_train)
y_pred = cloud_clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

logit = LogisticRegression()
logit.fit(X_train, y_train)
y_pred = logit.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
preds = pd.concat([pd.Series(logit.predict(X_test)), gee_test.reset_index(drop=True)],axis=1)
preds = preds.rename(columns={0: "pred"})
```
